```python
# this model is to create a class for the calories per day

class dailylog(object):
    def __init__(self, caloriesB, caloriesL, caloriesD):
        self.caloriesB = caloriesB
        self.caloriesL = caloriesL
        self.caloriesD = caloriesD
        
    def calorie_count(self, totalmaxcal = 1700):
        self.totalmaxcal = totalmaxcal
        self.caltoday = self.caloriesB + self.caloriesL + self.caloriesD
        self.caldiff = abs(self.caltoday - self.totalmaxcal)
        if self.caltoday > self.totalmaxcal:
            return("bluddclart, u need to cut and go gym my guy", self.caltoday, "you've gone over by", self.caldiff, "calories")
        else:
            return("nahh keep eatin u looking skimpy, you've only had", self.caltoday, "cals today")
    



userinput = input("please type in your calories for breakfast lunch and dinner (please seperate by commas):")
B,L,D = userinput.split(",")

Day1 = dailylog(int(B), int(L), int(D))
print(Day1.calorie_count())
```

```python
## this model is to calculate the either force, mass or acceleration when two of the three are given


class newtons2ndlaw(object):
    def __init__(self, force = None, mass = None, acceleration = None):
        self.force = force
        self.mass = mass
        self.acceleration = acceleration
    def eqns(self):
        if self.force is None:
            self.force = self.mass*self.acceleration
        elif self.mass is None:
             self.mass = self.force/self.acceleration
        elif self.acceleration is None:
            self.acceleration = self.force/self.mass
        return(self.force, self.mass, self.acceleration)
        
missing = input("what do you want to calculate (force/mass/acceleration):")
userinput = input("please add the values of the others")

val1, val2 = userinput.split(",")



if missing == "force":
    Equation1 = newtons2ndlaw(force = None, mass = float(val1), acceleration = float(val2))
    print("the force is:", Equation1.eqns()[0])
elif missing == "mass":
    Equation1 = newtons2ndlaw(force = float(val1), mass = None, acceleration = float(val2))
    print("the mass is:", Equation1.eqns()[1])
elif missing == "acceleration":
    Equation1 = newtons2ndlaw(force = float(val1), mass = float(val2), acceleration = None)
    print("the acceleration is:", Equation1.eqns()[2])
    


```
